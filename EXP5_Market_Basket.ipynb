```python
#importing libraries
import numpy as np
import pandas as pd

from apyori import apriori
```

```python
data=pd.read_csv("/content/Groceries_dataset.csv",names=["itemDescription"],header=None)

```

```python
data
```

```python
data.values
```

```python
data1=list(data["itemDescription"].apply(lambda x:x.split(',')))
```

```python
data1
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_data=te.fit(data1).transform(data1)
df=pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
#importing the apriori library
from mlxtend.frequent_patterns import apriori
```

```python
df1=apriori(df,min_support=0.01,use_colnames=True)
df1
```

```python
df1.sort_values(by="support",ascending=False)
```

```python
df1["length"]=df1["itemsets"].apply(lambda x:len(x))
df1
```

```python
df1[(df1["length"]==1) & (df1["support"]>0.05)]
```
